```python
import numpy as np
import matplotlib.pyplot as plt

fig,ax =plt.subplots(2,3,tight_layout=True)
x = [0,1,2,3]
y =[1,2,-5,2]
ax[0,0].plot(x,y)

x = np.linspace(0,10,100)
y = 2 + 2 * np.sin(2 * x)

ax[0,1].plot(x,y,linewidth=2.0)

ax[0,1].set(xlim=(0,10), xticks=np.arange(0,10),
       ylim=(-1,5), yticks=np.arange(-1,6))

x =np.random.normal(15,5,2000)
ax[0,2].hist(x)

np.random.seed(3)
x = 4 + np.random.normal(0,2,60)
y = 4 + np.random.normal(0,2,len(x))
ax[1,0].scatter(x,y)

plt.show()


```
